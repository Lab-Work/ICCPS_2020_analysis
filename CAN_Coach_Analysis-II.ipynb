```python
import bagpy
from bagpy import bagreader
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
import numpy as np
from strym import strymread

import warnings
warnings.filterwarnings("ignore")
import ntpath
```

```python
DataFolder = "../../PandaData/CanCoachExperiment/"
import glob
csvlist = glob.glob(DataFolder+"*.csv")
```

```python
r = strymread(csvlist[0])
```

```python
statespace = r.state_space(rate = 20)
```

```python
statespace
```

```python
fig, ax = strymread.create_fig(statespace.columns.shape[0]-1)
i = 0
for c in statespace.columns:
    if c == 'Time':
        continue
    ax[i].scatter(x = 'Time', y = c, data= statespace, s= 2, c= 'Time', cmap = strymread.sunset)
    ax[i].set_title(c, fontsize = 22)
    ax[i].set_xlabel('Time', fontsize =20)
    ax[i].set_ylabel(c, fontsize = 20)
    i = i + 1

fig.suptitle(ntpath.basename(csvlist[0]), y = 1.001, fontsize = 26)
fig.tight_layout()
fig.show()
```
